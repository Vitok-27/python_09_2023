# **Tоп пять самых популярных артистов**


```python
top_five_artists= df.groupby("artist(s)_name").count().sort_values(by="track_name", ascending=False)["track_name"][:5]
top_five_artists
```

```python
title_dict = {'fontsize': 24, 'fontweight': 'bold', 'color': 'black', 'family': 'sans-serif'}
lable_dict = {'fontsize': 20, 'color': 'red', 'family': 'sans-serif'}
```

```python
plt.figure(figsize=(10, 5))
top_five_artists.plot.barh()
plt.title("Топ пять самых популярных артистов", fontdict=title_dict)
plt.ylabel("Имя артиста", fontdict=lable_dict)
plt.show()
```

# **Вывод:**
## *На графике представленны топ пять самых популярных артистов начиная с 1930 по  2023.*




# **Топ пять самых танцевальных трека**

```python
top_five_danceability=df[["track_name","danceability_%"]].sort_values(by="danceability_%",ascending=False)[:5]
top_five_danceability
```

```python
title_dict = {'fontsize': 24, 'fontweight': 'bold', 'color': 'black', 'family': 'sans-serif'}
lable_dict = {'fontsize': 20, 'color': 'red', 'family': 'sans-serif'}
sns.set(rc={'figure.figsize': (10, 5)})
sns.barplot(x="danceability_%", y="track_name", data=top_five_danceability, palette="colorblind", alpha=0.3)
plt.title("Топ пять самых танцевальных трека", fontdict=title_dict)
plt.xlabel("danceability_%", fontdict=lable_dict)
plt.ylabel("Название трека", fontdict=lable_dict)
plt.show()
```

# **Вывод:**
## *На графике представленны топ пять самых тонцевальных треков начиная с 1930 по  2023.*




# **Артист с самой танцевальной песней**

```python
top_five_artists_danceability=df[["danceability_%", "track_name","artist(s)_name"]].sort_values(by="danceability_%", ascending=False)[:10]
top_five_artists_danceability
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability_%", y="artist(s)_name", data=top_five_artists_danceability)
plt.title("Артист с самой танцевальной песней", fontdict=title_dict)
plt.show
```

```python
title_dict = {'fontsize': 24, 'fontweight': 'bold', 'color': 'Red', 'family': 'sans-serif'}
plt.figure(figsize=(12,7))
plt.pie(x="danceability_%", data=top_five_artists_danceability, autopct="%1.1f%%", labels=top_five_artists_danceability.track_name)
plt.title("Самая танцевальная песня",fontdict=title_dict)
plt.show()
```

Вывод:
Артист с самым танцевальным треком c 1930 - 2023 стал Ed Sheeren
